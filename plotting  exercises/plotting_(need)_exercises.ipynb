# Exercise 1

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def my_function(x):
    return np.exp(-x / 10) * np.sin(x)

# Generate x values
x_values = np.linspace(0, 50, 1000)  # Adjust the range as needed

# Calculate corresponding y values
y_values = my_function(x_values)

# Plot the function with specified settings
plt.plot(x_values, y_values, label=r'$y = e^{-x/10} \sin(x)$', color='blue')  # Change the line color if needed
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title('$y = e^{-x/10} \sin(x)$', fontsize=20)

# Adjust the y-axis limit
plt.ylim([-1, 1])

# Add a grid with an opacity of 25%
plt.grid(True, alpha=0.25)

# Show the plot
plt.show()
```
