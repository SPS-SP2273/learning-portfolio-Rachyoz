```python
month_mapping = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

#file names right now: oct-08_singapore.txt
#change the filenames to 10-08-singapore.txt

def convert_filename(filename):
    parts = filename.split('_') #['oct-08', 'singapore.txt']
    if len(parts) == 2 and parts[1].endswith('.txt'):
        month, day_city = parts[0].split('-') #print(month,day_city) - oct 08
        
        
        # Convert month to numerical format
        month_numeric = month_mapping.get(month.lower())
        
        if month_numeric:
            new_filename = f'{month_numeric}-{day_city}-{parts[1]}'
            return new_filename
        else:
            print(f"Error: Invalid month '{month}' in filename.")
            return None
    else:
        print("Error: Invalid filename format.")
        return None

# Example usage
filename = 'oct-08_singapore.txt'
converted_filename = convert_filename(filename)

if converted_filename:
    print(f"Original filename: {filename}")
    print(f"Converted filename: {converted_filename}")
```
